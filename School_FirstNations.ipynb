## Attendance

### Indigenous and Non-Indigenous Attendance Rates at Mossman State High School (MSHS)

- H0: There is no significant difference between Indigenous and Non-Indigenous attendance rates at MSHS.
- H1: There is a significant difference between Indigenous and Non-Indigenous attendance rates at MSHS.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

# Load the data file
file_path = 'Data/School/Mossman State High School - Attendance by Year and Indigenous Status.xlsx'
attendance_data = pd.read_excel(file_path)

# Set plot style
sns.set(style="whitegrid")

# Create a plot to compare attendance rates
plt.figure(figsize=(12, 6))
sns.lineplot(data=attendance_data, x='Year', y='Attendance - Indigenous students (Percent)', label='Indigenous', marker='o')
sns.lineplot(data=attendance_data, x='Year', y='Attendance - Non-Indigenous students (Percent)', label='Non-Indigenous', marker='o')

plt.title('Attendance Rates at Mossman State High School (2014-2023)')
plt.ylabel('Attendance Rate (%)')
plt.xlabel('Year')
plt.legend(title='Group')
plt.tight_layout()
plt.show()
```

Visual Analysis - There is a consistent gap between the attendance rates of Indigenous and Non-Indigenous students at Mossman State High School over the years, with Non-Indigenous students consistently having higher attendance rates.

```python
# Extract attendance rates for Indigenous and Non-Indigenous students
indigenous_attendance = attendance_data['Attendance - Indigenous students (Percent)'].dropna()
non_indigenous_attendance = attendance_data['Attendance - Non-Indigenous students (Percent)'].dropna()

# Conduct the Mann-Whitney U Test
u_statistic, p_value = mannwhitneyu(indigenous_attendance, non_indigenous_attendance, alternative='two-sided')

u_statistic, p_value
```

Results: A Mann-Whitney U test was conducted to compare the attendance rates of Indigenous and Non-Indigenous students at Mossman State High School from 2014 to 2023. The results showed a significant difference in the attendance rates between the two groups, U = 0.0, p < .001, suggesting that Non-Indigenous students attended more frequently than their Indigenous counterparts.

Conclusion: The hypothesis test supports rejecting the null hypothesis (H0), which posited no significant difference in attendance rates between Indigenous and Non-Indigenous students. The evidence suggests that a significant difference does exist, affirming the alternative hypothesis (H1).

## Relationship between Indigenous Status and Attendance Rates across Far North Queensland High Schools

- H0: There is no significant relationship between Indigenous status and attendance rates across Far North Queensland High Schools (years 7 to 12).
- H1: There is a significant relationship between Indigenous status and attendance rates across Far North Queensland High Schools (years 7 to 12).


```python
# Load the Far North Queensland State Schools data
file_path_fnq = 'Data/School/Far North Queensland State Schools - Attendance Rates and SDA by Year Level.xlsx'
fnq_data = pd.read_excel(file_path_fnq)

# Apply filtering for grades 7 to 12 and years 2015 to 2019
fnq_correct_filtered_data = fnq_data[
    (fnq_data['Year'].between(2015, 2019)) &
    (fnq_data['Year Level'].isin(range(7, 13)))
]

# Set plot style
sns.set(style="whitegrid")

# Create a plot to compare attendance rates for Indigenous and Non-Indigenous students
plt.figure(figsize=(14, 7))
sns.lineplot(data=fnq_correct_filtered_data, x='Year', y='Attendance - Indigenous (Percent)', label='Indigenous', marker='o')
sns.lineplot(data=fnq_correct_filtered_data, x='Year', y='Attendance - Non-Indigenous (Percent)', label='Non-Indigenous', marker='o')

plt.title('Attendance Rates Across Far North Queensland High Schools (2015-2019)')
plt.ylabel('Attendance Rate (%)')
plt.xlabel('Year')
plt.legend(title='Group')
plt.tight_layout()
plt.show()
```

Visual Analysis - There is a noticeable gap in attendance rates between Indigenous and Non-Indigenous students in Far North Queensland High Schools from 2015 to 2019.

```python
# Extract attendance rates for Indigenous and Non-Indigenous students
indigenous_attendance_fnq = fnq_correct_filtered_data['Attendance - Indigenous (Percent)'].dropna()
non_indigenous_attendance_fnq = fnq_correct_filtered_data['Attendance - Non-Indigenous (Percent)'].dropna()

# Conduct the Mann-Whitney U Test
u_stat_fnq, p_value_fnq = mannwhitneyu(indigenous_attendance_fnq, non_indigenous_attendance_fnq, alternative='two-sided')

u_stat_fnq, p_value_fnq
```

Results: A Mann-Whitney U test was conducted to examine the relationship between Indigenous status and attendance rates in Far North Queensland High Schools for grades 7 to 12 during the years 2015 to 2019. The results revealed a significant difference between the attendance rates of Indigenous and Non-Indigenous students, U = 0.0, p < .001.

Conclusion: The hypothesis testing supports rejecting the null hypothesis (H0), which asserted no significant relationship between Indigenous status and attendance rates. The evidence suggests a significant relationship, confirming the alternative hypothesis (H1).

## Attendance rates by Indigenous status across different schools (Mossman State High School, Far North Queensland High Schools, and Queensland High Schools)

For Indigenous Students
- H: There is no significant difference in attendance rates among Indigenous students across Mossman State High School, Far North Queensland High Schools, and Queensland High Schools.
- H: There is a significant difference in attendance rates among Indigenous students across Mossman State High School, Far North Queensland High Schools, and Queensland High Schools.

For Non-Indigenous Students
- H: There is no significant difference in attendance rates among Non-Indigenous students across Mossman State High School, Far North Queensland High Schools, and Queensland High Schools.
- H: There is a significant difference in attendance rates among Non-Indigenous students across Mossman State High School, Far North Queensland High Schools, and Queensland High Schools.

```python
from scipy.stats import f_oneway

# Load data
mossman_attendance = pd.read_excel("Data/School/Mossman State High School - Attendance by Year and Indigenous Status.xlsx")
fnq_schools_attendance = pd.read_excel("Data/School/Far North Queensland State Schools - Attendance Rates and SDA by Year Level.xlsx")
qld_schools_attendance = pd.read_excel("Data/School/Queensland State Schools - Attendance Rates and SDA by Year Level.xlsx")

# Filter data for years 2015 to 2019
mossman_attendance_filtered = mossman_attendance[mossman_attendance['Year'].between(2015, 2019)]
fnq_schools_attendance_filtered = fnq_schools_attendance[fnq_schools_attendance['Year'].between(2015, 2019)]
qld_schools_attendance_filtered = qld_schools_attendance[qld_schools_attendance['Year'].between(2015, 2019)]

# Aggregate data to get mean attendance rates by year for Indigenous and Non-Indigenous groups
mossman_indigenous = mossman_attendance_filtered['Attendance - Indigenous students (Percent)'].groupby(mossman_attendance_filtered['Year']).mean()
mossman_non_indigenous = mossman_attendance_filtered['Attendance - Non-Indigenous students (Percent)'].groupby(mossman_attendance_filtered['Year']).mean()

fnq_indigenous = fnq_schools_attendance_filtered['Attendance - Indigenous (Percent)'].groupby(fnq_schools_attendance_filtered['Year']).mean()
fnq_non_indigenous = fnq_schools_attendance_filtered['Attendance - Non-Indigenous (Percent)'].groupby(fnq_schools_attendance_filtered['Year']).mean()

qld_indigenous = qld_schools_attendance_filtered['Attendance - Indigenous (Percent)'].groupby(qld_schools_attendance_filtered['Year']).mean()
qld_non_indigenous = qld_schools_attendance_filtered['Attendance - Non-Indigenous (Percent)'].groupby(qld_schools_attendance_filtered['Year']).mean()

# Perform ANOVA for Indigenous and Non-Indigenous separately
anova_indigenous = f_oneway(mossman_indigenous, fnq_indigenous, qld_indigenous)
anova_non_indigenous = f_oneway(mossman_non_indigenous, fnq_non_indigenous, qld_non_indigenous)

anova_indigenous, anova_non_indigenous
```

ANOVA Test for Attendance Rates by Indigenous Status
1. Indigenous Students:
    - F-statistic: 60.01
    - p-value: 0.000000564
    - This result indicates a statistically significant difference in the attendance rates of Indigenous students among Mossman State High School, Far North Queensland High Schools, and Queensland High Schools.
2. Non-Indigenous Students:
    - F-statistic: 0.155
    - p-value: 0.858
    - This result suggests that there are no statistically significant differences in the attendance rates of Non-Indigenous students among these schools.

In examining the relationship between Indigenous status and attendance rates among students at Mossman State High School, Far North Queensland High Schools, and Queensland High Schools from 2015 to 2019, a one-way ANOVA revealed a significant effect of school group on attendance rates for Indigenous students, F(2, 12) = 60.01, p < .001. This indicates that the attendance rates for Indigenous students significantly vary across different school groups. Conversely, the same analysis for Non-Indigenous students showed no significant differences in attendance rates, F(2, 12) = 0.155, p = .858, suggesting consistent attendance rates across school groups for Non-Indigenous students.

```python
# Create a DataFrame for plotting
data = {
    'Year': list(mossman_indigenous.index) * 6,
    'Attendance Rate': list(mossman_indigenous) + list(fnq_indigenous) + list(qld_indigenous) +
                       list(mossman_non_indigenous) + list(fnq_non_indigenous) + list(qld_non_indigenous),
    'Group': ['Mossman Indigenous'] * 5 + ['FNQ Indigenous'] * 5 + ['QLD Indigenous'] * 5 +
             ['Mossman Non-Indigenous'] * 5 + ['FNQ Non-Indigenous'] * 5 + ['QLD Non-Indigenous'] * 5
}
df_plot = pd.DataFrame(data)

# Adjust the 'dashes' parameter correctly by defining specific styles for each group
dashes = {'Mossman Indigenous': (2, 2), 'FNQ Indigenous': (2, 2), 'QLD Indigenous': (2, 2),
          'Mossman Non-Indigenous': '', 'FNQ Non-Indigenous': '', 'QLD Non-Indigenous': ''}

# Adjust the color palette so that the same regions have the same color
palette = {'Mossman Indigenous': 'blue', 'FNQ Indigenous': 'green', 'QLD Indigenous': 'red',
           'Mossman Non-Indigenous': 'blue', 'FNQ Non-Indigenous': 'green', 'QLD Non-Indigenous': 'red'}

# Plotting with corrected color and line settings
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='Attendance Rate', hue='Group', style='Group', markers=True, dashes=dashes, palette=palette, data=df_plot)
plt.title('Attendance Rates by Indigenous Status (2015-2019)')
plt.xlabel('Year')
plt.ylabel('Average Attendance Rate (%)')
plt.legend(title='Group')
plt.grid(True)
plt.show()
```

## Trend in the Gap between Indigenous and Non-Indigenous Attendance Rates at MSHS over Time

- H0: There is no significant trend in the gap between Indigenous and Non-Indigenous attendance rates at MSHS over time.
- H1: There is a significant trend in the gap between Indigenous and Non-Indigenous attendance rates at MSHS over time.


```python
# Load the data
file_path = 'Data/School/Mossman State High School - Attendance by Year and Indigenous Status.xlsx'
attendance_data = pd.read_excel(file_path)

import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the gap in attendance rates between Indigenous and Non-Indigenous students
attendance_data['Gap'] = attendance_data['Attendance - Non-Indigenous students (Percent)'] - attendance_data['Attendance - Indigenous students (Percent)']

# Remove rows with NaN values in the 'Gap' column
cleaned_data = attendance_data.dropna(subset=['Gap'])

# Run the regression analysis with cleaned data
X_cleaned = sm.add_constant(cleaned_data['Year'])  # adding a constant for intercept
Y_cleaned = cleaned_data['Gap']

# Perform linear regression on the cleaned data
model_cleaned = sm.OLS(Y_cleaned, X_cleaned).fit()

# Display the visualization with cleaned data
plt.figure(figsize=(10, 6))
sns.regplot(x='Year', y='Gap', data=cleaned_data, ci=None, marker='o', line_kws={'color':'red'})
plt.title('Trend in the Gap between Indigenous and Non-Indigenous Attendance Rates at MSHS')
plt.xlabel('Year')
plt.ylabel('Attendance Rate Gap (%)')
plt.grid(True)
plt.show()

```

Regression Results:
- The regression equation is: Gap = 1680.7917 - 0.8250 * Year
- R2 = 0.801 suggests that approximately 80.1% of the variability in the attendance gap can be explained by the linear model.
- The slope of the line ( coefficient for the Year) is -0.8250, with a p-value of 0.001, which is statistically significant (p < 0.05).

Results:
A linear regression analysis was conducted to investigate the presence of a trend in the attendance rate gap between Indigenous and Non-Indigenous students at Mossman State High School from 2014 to 2023. The results of the regression indicated that the year significantly predicted the attendance rate gap,  = -0.825, t(7) =0.825,t(7)=5.304,p=.001. The model explained 80.1% of the variance (R2 =.801), demonstrating a significant decrease in the gap over time. This analysis provides strong evidence to reject the null hypothesis (H0) and accept the alternative hypothesis (H1), indicating a significant trend in the gap between Indigenous and Non-Indigenous attendance rates at Mossman State High School over the analyzed period.

The plot above illustrates the decreasing trend in the attendance rate gap, highlighting a gradual improvement in attendance parity between the two student groups.

## SDA

### Difference in Disciplinary Absence Rates between Indigenous and Non-Indigenous Students at MSHS

- H0: There is no significant difference in disciplinary absence rates between Indigenous and Non-Indigenous students at MSHS.
- H1: There is a significant difference in disciplinary absence rates between Indigenous and Non-Indigenous students at MSHS.

```python
# Adjust the filter to account for both string and numeric representations of year levels
high_school_levels = [7, 8, 9, 10, 11, 12, '7', '8', '9', '10', '11', '12']
indigenous_cols = [
    'Disciplinary Absences - Short Suspension - Indigenous',
    'Disciplinary Absences - Long Suspension - Indigenous',
    'Disciplinary Absences - Exclusion - Indigenous',
    'Disciplinary Absences - Cancellation - Indigenous',
    'Disciplinary Absences - Charge Suspension - Indigenous'
]

non_indigenous_cols = [
    'Disciplinary Absences - Short Suspension - Non-Indigenous',
    'Disciplinary Absences - Long Suspension - Non-Indigenous',
    'Disciplinary Absences - Exclusion - Non-Indigenous',
    'Disciplinary Absences - Cancellation - Non-Indigenous',
    'Disciplinary Absences - Charge Suspension - Non-Indigenous'
]
# Filter the data for the relevant year levels
high_school_data_corrected = fnq_data[fnq_data['Year Level'].isin(high_school_levels)]

# Calculate total disciplinary absences for Indigenous and Non-Indigenous students
total_disciplinary_absences_indigenous_corrected = high_school_data_corrected[indigenous_cols].sum(axis=1)
total_disciplinary_absences_non_indigenous_corrected = high_school_data_corrected[non_indigenous_cols].sum(axis=1)

# Run the Mann-Whitney U test
stat_corrected, p_value_corrected = mannwhitneyu(
    total_disciplinary_absences_indigenous_corrected,
    total_disciplinary_absences_non_indigenous_corrected,
    alternative='two-sided'
)

stat_corrected, p_value_corrected

# Calculate total disciplinary absences for both groups
total_indigenous = total_disciplinary_absences_indigenous_corrected.sum()
total_non_indigenous = total_disciplinary_absences_non_indigenous_corrected.sum()

# Prepare data for the bar graph
totals_data = pd.DataFrame({
    "Group": ["Indigenous", "Non-Indigenous"],
    "Total Disciplinary Absences": [total_indigenous, total_non_indigenous]
})

# Plotting the bar graph
plt.figure(figsize=(8, 6))
sns.barplot(x='Group', y='Total Disciplinary Absences', data=totals_data)
plt.title('Total Disciplinary Absences in Far North Queensland High Schools')
plt.ylabel('Total Disciplinary Absences')
plt.xlabel('Student Group')
plt.show()
```

The Mann-Whitney U test on the total disciplinary absences between Indigenous and Non-Indigenous high school students in Far North Queensland results in a U-statistic of 1489.0 with a p-value of 0.0137.

Based on this result, we reject the null hypothesis (H0) that there is no significant difference in total disciplinary absences between Indigenous and Non-Indigenous students. The p-value suggests there is a statistically significant difference between the two groups.

## Difference in Disciplinary Absence Rates between Indigenous and Non-Indigenous Students for Far North Queensland

- H0: There is no significant difference in the rate of disciplinary absences (total of all types) between Indigenous and Non-Indigenous highschool students across year levels in Far North Queensland.
- H1: There is a significant difference in the rate of disciplinary absences (total of all types) between Indigenous and Non-Indigenous highschool students across year levels in Far North Queensland.

## 

```python
enrolment_data_path = 'Data/School/Far North Queensland - Enrolment Numbers.xlsx'
absence_data_path = 'Data/School/Far North Queensland State Schools - Attendance Rates and SDA - Aggregated Data for High School.xlsx'

# Read the data
enrolment_data = pd.read_excel(enrolment_data_path)
absence_data = pd.read_excel(absence_data_path)

# Combine the total disciplinary absences for Indigenous and Non-Indigenous students
absence_data['Total Disciplinary Absences - Indigenous'] = (
    absence_data['Disciplinary Absences - Short Suspension - Indigenous'] +
    absence_data['Disciplinary Absences - Long Suspension - Indigenous'] +
    absence_data['Disciplinary Absences - Cancellation - Indigenous'] +
    absence_data['Disciplinary Absences - Charge Suspension - Indigenous']
)

absence_data['Total Disciplinary Absences - Non-Indigenous'] = (
    absence_data['Disciplinary Absences - Short Suspension - Non-Indigenous'] +
    absence_data['Disciplinary Absences - Long Suspension - Non-Indigenous'] +
    absence_data['Disciplinary Absences - Cancellation - Non-Indigenous'] +
    absence_data['Disciplinary Absences - Charge Suspension - Non-Indigenous']
)

# Merge enrolment data to compute rates per 1000 students
combined_data = pd.merge(absence_data, enrolment_data, on='Year', how='left')

# Calculate the rates per 1000 students
combined_data['Rate per 1000 - Indigenous'] = (
    combined_data['Total Disciplinary Absences - Indigenous'] / combined_data['Number of Enrolled Students that Are Indigenous'] * 1000
)

combined_data['Rate per 1000 - Non-Indigenous'] = (
    combined_data['Total Disciplinary Absences - Non-Indigenous'] / (combined_data['Enrolment total'] - combined_data['Number of Enrolled Students that Are Indigenous']) * 1000
)

# Select relevant columns for analysis
analysis_data = combined_data[['Year', 'Rate per 1000 - Indigenous', 'Rate per 1000 - Non-Indigenous']]
analysis_data

# Extract the rates for each group
rates_indigenous = analysis_data['Rate per 1000 - Indigenous']
rates_non_indigenous = analysis_data['Rate per 1000 - Non-Indigenous']

# Perform the Mann-Whitney U test
u_statistic, p_value = mannwhitneyu(rates_indigenous, rates_non_indigenous, alternative='two-sided')

u_statistic, p_value

# Create a line plot for the rates
plt.figure(figsize=(10, 6))
sns.lineplot(data=analysis_data, x='Year', y='Rate per 1000 - Indigenous', marker='o', label='Indigenous')
sns.lineplot(data=analysis_data, x='Year', y='Rate per 1000 - Non-Indigenous', marker='o', label='Non-Indigenous')
plt.title('Disciplinary Absence Rates per 1000 Students by Group')
plt.ylabel('Rate per 1000 Students')
plt.xlabel('Year')
plt.legend(title='Group')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create a bar graph for the rates
plt.figure(figsize=(12, 8))
sns.barplot(data=analysis_data.melt(id_vars='Year', var_name='Group', value_name='Rate per 1000'),
             x='Year', y='Rate per 1000', hue='Group')
plt.title('Disciplinary Absence Rates per 1000 Students by Group')
plt.ylabel('Rate per 1000 Students')
plt.xlabel('Year')
plt.legend(title='Group', labels=['Indigenous', 'Non-Indigenous'])
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

A Mann-Whitney U test revealed a significant difference (U = 64.0, p = .00016, two-tailed). The results suggest that the rates of disciplinary absences are significantly different between these groups, leading to the rejection of the null hypothesis.
